# Importing Packages

```python
import pybel
from pybel.io.jupyter import to_jupyter
from utils import *
```

# Import cached KG 

```python
infile = open('data/export/Monkeypox_KG_15jan.pkl','rb')
mpox_graph = pickle.load(infile)
infile.close()
```

# Get annotations from protein or chemicals

```python
#list entities of interest separated by a space
entityList = 'CHEMBL1643 CHEMBL152 CYP3A4 NS5B pol'.split(' ')

prtn_attr = nx.get_node_attributes(mpox_graph,'UniProt')
chem_attr = nx.get_node_attributes(mpox_graph,'ChEMBL')

for node in mpox_graph.nodes():
    unique_node = []
    
    if node.name in entityList:
        if node not in unique_node:
            
            unique_node.append(node)
            print(node.name)
       
            if isinstance(node,pybel.dsl.Abundance):
                print(chem_attr[node])
                
            if isinstance(node,pybel.dsl.Protein):
                print(prtn_attr[node])
            
```

# Get annotations from protein-pathway relationships

```python
for u,v,t in mpox_graph.edges(data=True):
    if 'annotation' in t:
        if 'Reactome' in t['annotation']:
            print(u,'association',v)
            print(t['annotation']['Reactome'])
```
